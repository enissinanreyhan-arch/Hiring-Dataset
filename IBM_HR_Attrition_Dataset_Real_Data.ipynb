NORMATIVE EFFECT EXAMPLE

As per the Caselaw of the ECJ (CK vs. Magistrat der Stadt Wien & Dun & Bradstreeet Austria GmBH), it is clarified that the data controller do not need to provide detailed information about the algorithm and the court emphasizes on contractual examples.

Thus, some guidance has been provided for ECJ to establish legal requirements for explanations to be provided, besides the technical ones. As LLMs are used as external tools for decision making as well as for explanations, it is also included in the workbook.

Within the scope of this workbook, a supervised learning xAI tool (DICE) is run and the counterfactual examples are examined from legal perspective and in line with Caselaw of ECJ.

LLM is used to explain the counterfactual examples generated by the xAI tool. Purpose is to show how Counterfactual Examples are created by Dice algorithm and explained by LLM Model (OpeanAI).

Case: Model outcome is used for performance assessment, with an automated formula. Employees with attrition risk are subject to a disadvantaged coefficient, for wage increase.

NORMATIVITY RULE

A policy rule added, with two approaches:

a) Employees above the probability threshold (e.g., 0.50) are penalized with less salary increase (as they are more inclined to quit, so less investment is done.)

IF AttritionRiskScore > 0.50
THEN SalaryIncreaseMultiplier = 0.7
ELSE SalaryIncreaseMultiplier = 1.0

b) No threshold is set, but the bonus/salary increase is penalised with the probability.

PerformanceScore = HumanScore × (1 − α·AttritionRisk)

First imposes a decision boundary, and a direct rule. Thus, counterfactuals can be generated. For the second, the counterfactual example is not meaningful, as the algorithm's effect is continous.

This shows how the outcome of the algorithm can affect individuals' rights. If no real human intervention is in the loop, the algorithms falls within the scope of the Article 22 of the GDPR.

As the algorithmically generated norm directly affects individual incentives, it may give rise to a normative feedback loop. Reduced incentives can lead to lower motivation and performance, thereby increasing the likelihood of attrition. In this way, the model’s prediction does not merely describe future behavior but actively contributes to producing it.



```python

m = dice_ml.Model(model=model, backend="sklearn")
d = dice_ml.Data(dataframe=train_dataset, continuous_features=continuous_cols, outcome_name='Attrition')
exp = dice_ml.Dice(d, m, method="random")
```

In this case, counterfactual examples are problematic from legal side. The counterfactual examples are not useful for individuals, however, carries real risk to be discriminative (age, gender, marital status etc). Additionally, there are some variables that cannot be easily or logically changed (DistanceFromHome, JobRole, NumCompaniesWorked, Overtime, YearsAtCompany). Thus, at least discriminative variables should be excluded.

```python
e1 = exp.generate_counterfactuals(
    query_instance,
    total_CFs=4,
    desired_class=0,
    stopping_threshold=0.8,
    features_to_vary="all")
e1.visualize_as_dataframe(show_only_changes=True)
```

```python
def show_final(cf_e):
  cf_dataframe = cf_e.cf_examples_list[0].final_cfs_df
  cf_df = e1.cf_examples_list[0].final_cfs_df.copy()
  true_label = y_test.loc[idx]
# ORIG instance
  orig_df = query_instance.copy()

  orig_df.index = ["ORIG"]
  cf_df.index = [f"CF{i+1}" for i in range(len(cf_df))]

  combined = pd.concat([orig_df, cf_df], axis=0)
  combined.loc["ORIG", "Attrition"] = true_label

  orig_row = combined.loc["ORIG"]

# Değişen kolonlar: en az bir CF, ORIG'den farklıysa
  changed_cols = [
    col for col in combined.columns
    if not (combined[col] == orig_row[col]).all()
  ]
  cf_changes_df = combined[changed_cols].copy()
  return cf_changes_df



```

ASKING LLM TO INDEPENDENTLY EVALUATE THE INSTANCE AND CREATE COUNTERFACTUALS

```python
# Target
target = "Attrition"

# Nominal kategorikler (raw halde: OneHotEncoder pipeline’da)
categorical_cols = [
    "BusinessTravel", "Department", "EducationField", "JobRole",
    "MaritalStatus", "Gender", "OverTime"
]

# Continuous (scale edilecek / modelde de anlamlı)
continuous_cols = [
    "Age","DailyRate","DistanceFromHome","HourlyRate","MonthlyIncome","MonthlyRate",
    "PercentSalaryHike","TotalWorkingYears","TrainingTimesLastYear","YearsAtCompany",
    "YearsInCurrentRole","YearsSinceLastPromotion","YearsWithCurrManager"
]

# Ordinal (scale etmiyoruz)
ordinal_cols = [
    "Education","EnvironmentSatisfaction","JobInvolvement","JobLevel","JobSatisfaction",
    "PerformanceRating","RelationshipSatisfaction","StockOptionLevel","WorkLifeBalance",
    "NumCompaniesWorked"
]

# Değiştirilemez / ayrımcılığa açık / "kişisel özellik" (CF’lerde kilit)
immutable_cols = [
    "Age", "Gender", "MaritalStatus"
]
```

```python
def instance_to_markdown_row(instance_row, features):
    """1-row Series -> markdown-friendly dict-like string pieces"""
    return {feat: instance_row[feat] for feat in features}

def instance_to_feature_string(instance_row, features):
    return ", ".join(f"{feat} = {instance_row[feat]}" for feat in features)
```

```python
def llm_attrition_risk_and_counterfactuals(
    client,
    employee_row,             # pandas Series (1 row)
    all_features,             # list of feature names in the raw (non-dummied) dataset
    categorical_cols,
    continuous_cols,
    ordinal_cols,
    immutable_cols,
    model_name="gpt-4.1-mini"
):
    employee_str = instance_to_feature_string(employee_row, all_features)

    # Hangi kolonların değiştirilebileceğini netleştir
    mutable_cols = [c for c in all_features if c not in immutable_cols]

    dataset_description = f"""
You are analyzing an employee attrition (leaving the company) risk scenario using an HR dataset.
Target: Attrition (1 = employee leaves, 0 = employee stays).

Feature groups:
- Continuous numeric: {continuous_cols}
- Ordinal (coded integers): {ordinal_cols}
- Categorical: {categorical_cols}

IMPORTANT CONSTRAINTS (Fairness / non-discrimination / immutability):
- You MUST NOT change these features in counterfactuals: {immutable_cols}
- Treat them as fixed (either legally sensitive or not realistically changeable).

Scenario:
Assume the company uses an automated system to estimate attrition risk and then decides the salary increase/bonus.
Higher predicted attrition risk may lead to LOWER salary increase (a potentially harmful incentive).
Your job is to provide a human-readable risk assessment and propose realistic, actionable changes
that could plausibly reduce attrition risk while respecting the constraints.
""".strip()

    user_prompt = f"""
{dataset_description}

Current employee instance:
{employee_str}

TASKS:
1) Provide:
   - A short decision: "Higher attrition risk" OR "Lower attrition risk"
   - An estimated attrition probability between 0.0 and 1.0 (subjective estimate)
   - 2-4 bullet points explaining the main risk drivers (based on the given features)

2) Generate EXACTLY 4 counterfactual examples (CF1..CF4) that could reduce attrition risk.
   - Do NOT change immutable features: {immutable_cols}
   - Keep changes small and realistic, focusing on workplace/policy-related factors when possible.
   - Use only the existing feature columns (no new columns).
   - For any categorical feature you change, pick ONE valid alternative category value.

OUTPUT FORMAT (Markdown):
A) First: a short paragraph with the decision and probability.
B) Then: a Markdown table with 5 rows total:
   - Row 1 must be the ORIGINAL instance (label it "ORIG")
   - Rows 2-5 must be CF1, CF2, CF3, CF4
Columns must be:
| ID | (all features...) | Rationale |

- Put feature values in the same columns as the features.
- Rationale: one short sentence describing why risk should be lower.
- Keep the table compact; don’t add extra sections.

Make sure the output is valid Markdown. Modified variables should be in bold in the table.
""".strip()

    response = client.chat.completions.create(
        model=model_name,
        messages=[{"role": "user", "content": user_prompt}],
    )
    return response.choices[0].message.content
```

```python
# x_test: raw features DataFrame
# y_test: target Series (0/1)

idx = y_test[y_test == 1].index[0]     # ilk Attrition=1 örneği
query_instance = x_test.loc[[idx]]     # 1-row DataFrame
employee_row = query_instance.iloc[0]

all_features = list(x_test.columns)

out = llm_attrition_risk_and_counterfactuals(
    client=client,
    employee_row=employee_row,
    all_features=all_features,
    categorical_cols=categorical_cols,
    continuous_cols=continuous_cols,
    ordinal_cols=ordinal_cols,
    immutable_cols=immutable_cols,
    model_name="gpt-4.1-mini"
)

print(out)
```

CONCLUSION

The current use case, where the outcome of the algorithm is used as an input for automatic salary increase, falls under the scope of the article 22 and right to explanation.

As this is a classification algorithm, the deicision boundary is subject to change, in line with the algorithm's metrics like F1, recall, AUC, Precision and Accuracy. It is possible to provide counterfactual example for this case. There may be the case where probability generated by the algorithm is used as a variable for the calculation of the wage increase. In this case, no decision boundary is determined, thus, it is not possible to generate a counterfactual example.

Additionally, in this use case, the counterfactual examples are not meaningful for data subject. Although it gives certain grounds to contest to decision (the variables are discriminative or are not in employee's discretion etc.), it is possible to exclude the variables, which have major effect on the algorithm's decision, in the process of creating counterfactual examples.

However, counterfactual explanations are meaningful for company administration and may guide them on how to avoid their employees' exit from the company. Explanations created by LLM for these counterfactual examples are explanatory for the company administration, as well. As these counterfactual examples are instance based, it may provide information for specific employees.

However, this use case is an important example for algorithms with normative effect. So it must be subject to right to explanation and fall under the scope of GDPR. As counterfactual examples are not satisfactory, other techniques must be applied for data owner, to exercise his/her right to explanation.
